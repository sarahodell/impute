```python
def co_loc(f1,c=10):
    """
    Returns regions of a chromosome received from particular parents
    in an individual ril chromosome (int)
    Input:
        f1: pandas df of format: "start" "donor".
        Example: rils[1]"""
    locs=[]
    last = f1.iloc[0]
    counter = f1.iloc[0][0]
    for index,row in f1.iterrows():
        if row[1] != last[1]:
            start = int(row[0])-1
            end = int((row[0]))
            locs.append([c,counter,start,last[1]])
            counter=end
        last = row
    locs.append([c,counter,int((f1.iloc[-1][0])),last[1]])
    return locs
```

```python
locs = co_loc(breaks)
```

Developing a way of reading in Projection Alignment files and comparing them to the actual simulation breakpoint data

```python
from subprocess import Popen,PIPE
```

```python
donors={}
```

```python
ril_breakpoints=[]
```

```python
process = Popen(['awk','/#Donor Haplotypes/{flag=1;next}/#Taxa Breakpoints/{flag=0}flag','../B73xOh43_chr10_SimRILS_chr10_IMPUTED_ProjA.pa.txt'],stdout=PIPE,stderr=PIPE)
stdout,sterr=process.communicate()  
```

```python
for i in stdout.split('\n')[:-1]:
    line = i.split('\t')
    donors[line[0]]=line[1]
```

```python
donors
```

```python
process = Popen(['awk','/#Block are defined chr:startPos:endPos:donor1:donor2/{flag=1;next}/,0/{flag=0}flag','../B73xOh43_chr10_SimRILS_chr10_IMPUTED_ProjA.pa.txt'],stdout=PIPE,stderr=PIPE)
stdout,sterr=process.communicate()
```

```python
for i in stdout.split('\n')[:-1]:
    line = i.split('\t')
    brks = {}
    ril=line[0]
    for j in line[1:-1]:
        info=j.split(':')
        chrom=info[0]
        start=info[1]
        end=info[2]
        donor1=info[3]
        donor2=info[4]
        brks = {'ril':ril, 'chr':chrom, 'start':start, 'end':end, 'donor1': donors[donor1], 'donor2':donors[donor2]}
        ril_breakpoints.append(brks)
```

```python
metrics ={'ril':[],'perc_correct':[]}
```

```python
for r in proja['ril'].unique():
    pril = proja[proja['ril']==r]
    right=0
    total=0
    for index,row in pril.iterrows():
        pstart = int(row['start_pred'])
        pend = int(row['end_pred'])
        actual = rilsim[rilsim['ril']==row['ril']]
        for k in range(pstart,pend+1):
            for i,j in actual.iterrows():
                if int(j['start'])< k and int(j['end'])>=k:
                    inrange=j['donor']
                    if inrange==row['donor_pred']:
                        right+=1
            total+=1
    metrics['ril'].append(row['ril'])
    metrics['perc_correct'].append(round(float(right)/total,2))
```
